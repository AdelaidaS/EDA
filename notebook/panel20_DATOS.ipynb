```python
panel20_1.columns = ['ventas(miles €)', 'T.ESPAÑA', 'CATALUÑA', 'ARAGÓN',
       'ILLES BALEARS', 'COMUNITAT VALENCIANA', 'REGIÓN DE MURCIA',
       'ANDALUCÍA', 'COMUNIDAD DE MADRID', 'CASTILLA - LA MANCHA',
       'EXTREMADURA', 'CASTILLA Y LEÓN', 'GALICIA', 'PRINCIPADO DE ASTURIAS',
       'CANTABRIA', 'PAIS VASCO', 'LA RIOJA', 'C. FORAL DE NAVARRA',
       'CANARIAS']
```

```python
panel20_1
```

```python
panel20_1_tortilla = panel20_1[panel20_1["ventas(miles €)"].str.contains("tortilla|.TOTAL ALIMENTACION", case=False)]
panel20_1_tortilla
```

```python
totales_sum = panel20_1_tortilla.iloc[[1, 2]].sum()
#se añade el valor obtenido al df completo
panel20_kg_tortilla = panel20_1_tortilla.append(totales_sum, ignore_index=True)
panel20_kg = pd.DataFrame (panel20_kg_tortilla)
panel20_kg
```

```python
#datos de españa total y tortillas
panel20_España = panel20_kg.iloc[[0,3], 0:2]
panel20_España
```

```python
#datos de españa total y tortillas
panel20_CCAA = panel20_kg.drop(panel20_kg.columns[1], axis=1).iloc[[3], :]
panel20_CCAA
```

```python
# exporta el DataFrame 
panel20_kg.to_csv('/Users/Ade/src/data/panel20_kg.csv', index=False)
panel20_España.to_csv('/Users/Ade/src/data/panel20_España.csv', index=False)
panel20_CCAA.to_csv('/Users/Ade/src/data/panel20_CCAA.csv', index=False)
```
