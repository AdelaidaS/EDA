```python
panel21_canal= pd.read_excel ("2021datospanelconsumoalimentario.xlsx", sheet_name=3)
panel21_canal.head(3)
```

```python
#unificar el nombre de españa y nombrar la serie
panel21_canal.columns = ['ventas_canal', 'T.ESPAÑA', 'TDA.TRADICIONAL', 'ECONOMATO/COOP.',
       'HIPERMERCADOS', 'SUPER/AUTOS/G.ALM.', 'SUPER+AUTOS', 'DISCOUNTS',
       'TDA.CONGELADOS', 'HERBORISTERIA', 'FARMACIA', 'VTA.DOMICILIO',
       'PESCADERIA', 'CARNICERIA/CHARC.', 'MERCADOS/AMBULANT.',
       'MERCADOS Y PLAZAS', 'MERCADILLOS', 'LECHERIA', 'PANADERIA',
       'BAR-BODEGA', 'AUTOCONSUMO', 'COMPRA DIRECT.PRODUCTOR', 'RESTO',
       'INTERNET']
```

```python
#aqui extractado kg vendidos por canal y por total alimentacion y especifico
panel21_canal_torti = panel21_canal[panel21_canal["ventas_canal"].str.contains("tortilla", case=False)]
panel21_canal_torti
```

```python
#dar el dato por valor unico de tortillas
totales_sum = panel21_canal_torti.iloc[[0,1]].sum()
#se añade el valor obtenido al df completo
panel21_canal_tortilla = panel21_canal_torti.append(totales_sum, ignore_index=True).drop([0,1])
#hay canales vacios que distorsionan, eliminar
panel21_canal_tortilla.drop(['FARMACIA','MERCADOS/AMBULANT.','MERCADOS Y PLAZAS', "VTA.DOMICILIO", "AUTOCONSUMO", 'MERCADILLOS', 'LECHERIA','PANADERIA', 'BAR-BODEGA' ], axis=1, inplace=True)
panel21_canal_tortilla
panel21_canal_tortilla.replace({"TORTILLAS REFRIGERADTORTILLAS REFRIGE.SG": "ventas"}, inplace=True)
```

```python
df_transpuesto = panel21_canal_tortilla.transpose()
df_transpuesto = pd.DataFrame (df_transpuesto)
df_transpuesto
df_transpuesto['ventas_canal'] =df_transpuesto.iloc[:, 0] 
canal = df_transpuesto.drop(df_transpuesto.columns[1], axis=1)
canal
```

```python
#exportar a csv 
ruta_archivo = r"C:\Users\Ade\src\data\canal.csv"
canal.to_csv(ruta_archivo)
```
