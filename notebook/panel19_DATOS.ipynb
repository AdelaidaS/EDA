```python
import pandas as pd
from functools import reduce
import matplotlib.pyplot as plt
```

```python
ruta_archivo = r"\Users\Ade\src\data\2019datospanelconsumoalimentario.xlsx"
panel19_kg = pd.read_excel(ruta_archivo)
panel19_kg
```

```python
panel19_kg.columns = ['ventas(miles €)', 'ESPAÑA', 'CATALUÑA', 'ARAGÓN', 'ILLES BALEARS',
       'COMUNITAT VALENCIANA', 'REGIÓN DE MURCIA', 'ANDALUCÍA',
       'COMUNIDAD DE MADRID', 'CASTILLA - LA MANCHA', 'EXTREMADURA',
       'CASTILLA Y LEÓN', 'GALICIA', 'PRINCIPADO DE ASTURIAS', 'CANTABRIA',
       'PAIS VASCO', 'LA RIOJA', 'C. FORAL DE NAVARRA', 'CANARIAS']
```

```python
#aqui extractado kg vendidos por CCAA y por total alimentacion y especifico
panel19_kg_torti = panel19_kg[panel19_kg["ventas(miles €)"].str.contains("tortilla|.TOTAL ALIMENTACION", case=False)]
panel19_kg_torti
```

```python
# suma los valores individuales de los datos de los tipos de tortilla
totales_sum = panel19_kg_torti.iloc[[1, 2]].sum()
#se añade el valor obtenido al df completo
panel19_kg_tortilla = panel19_kg_torti.append(totales_sum, ignore_index=True)
panel19_kg_tortilla = pd.DataFrame (panel19_kg_tortilla)
panel19_kg_tortilla
```

```python
#datos de españa total y tortillas
panel19_España = panel19_kg_tortilla.iloc[[0,3], 0:2]
panel19_España
```

```python
#datos de españa total y tortillas
panel19_CCAA = panel19_kg_tortilla.drop(panel19_kg_tortilla.columns[1], axis=1).iloc[[3], :]
panel19_CCAA
```

```python
# exporta el DataFrame 
panel19_kg.to_csv('/Users/Ade/src/data/panel19_kg.csv', index=False)
panel19_España.to_csv('/Users/Ade/src/data/panel19_España.csv', index=False)
panel19_CCAA.to_csv('/Users/Ade/src/data/panel19_CCAA.csv', index=False)
```
